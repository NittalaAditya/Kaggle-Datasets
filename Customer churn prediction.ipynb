### Let's check for some outliers in the data & remove them

```python
def checkOutliers(customerDf,non_cat):
    for i in (non_cat):
        if (i!='churn'):
            sns.boxplot(customer['churn'],customer[i])
            plt.xlabel('Churn',fontsize=15)
            plt.ylabel(i,fontsize=15)
            plt.title('Boxplot Plot for {}'.format(i),fontsize=15)
            plt.show()
        
```

```python
checkOutliers(customer,non_cat)
```

There seems to be a lot of outliers everywhere so the net step should be to remove them

```python
def remove_outliers(customer,non_cat):
    for i in non_cat:
        if(i!='churn'):
            quartile1 = customer[i].quantile(0.25)
            quartile2 = customer[i].quantile(0.75)
            iqr = quartile2-quartile1
            upper_bound = quartile2+1.5*iqr
            lower_bound = quartile1-1.5*iqr
            customer[i] = customer[i].mask(customer[i]>upper_bound,customer[i].median())
            customer[i] = customer[i].mask(customer[i]<lower_bound,customer[i].median())
    return customer
            
```

```python
newDf = remove_outliers(customer,non_cat)
```

```python
#Check for outliers after the change as well
checkOutliers(newDf,non_cat)
```

### Handling Imbalanced Dataset

0 represent the no churn and 1 represent the churn so there are huge difference in the class. so we need to balanced the dataset
We have to use upsampling for handling the dataset

```python
x = encodingDf.drop(columns='churn',axis=1)
y = encodingDf['churn']
```

```python
from sklearn.model_selection import train_test_split
#id_submission = encoding_test.id
#encoding_test = encoding_test.drop(columns='id',axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from  imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(sampling_strategy='minority',random_state=42)
x_sm,y_sm = sm.fit_resample(x_train,y_train)
```

```python
y_sm.value_counts()
```

Scaling dataset

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x_train  = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

## It's time to build some models

Let's start with Support Vector Classification first

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import cohen_kappa_score
svc = SVC(kernel='rbf', decision_function_shape='ovr')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print('Accuracy: ')
print('{}'.format(accuracy_score(y_test, y_pred)))
print('Classification report: ')
print('{}'.format(classification_report(y_test, y_pred)))
print('Confusion Matrix')
print('{}'.format(confusion_matrix(y_test, y_pred)))
print('Cohen kappa score: ')
print('{}'.format(cohen_kappa_score(y_test, y_pred)))
```

Let's check accuracy in Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_rfc = rfc.predict(x_test)

print("Accuracy Score : {}".format(accuracy_score(y_test,y_rfc)))
print("Classification Report \n {}".format(classification_report(y_test,y_rfc)))
print("Confusion Matrix \n {}".format(confusion_matrix(y_test,y_rfc)))
print('Cohen kappa score: ')
print('{}'.format(cohen_kappa_score(y_test, y_rfc)))
```
